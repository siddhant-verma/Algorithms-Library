```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Loading an example dataset from Seaborn

```python
tips = sns.load_dataset("tips")
```

First five rows

```python
print(tips.head())
```

Example 1: Scatterplot

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", style="sex")
plt.title("Scatterplot Example from Seaborn Docs")
plt.show()
```

Example 2: Histogram

```python
sns.histplot(data=tips, x="total_bill", kde=True, hue="sex")
plt.title("Histogram Example from Seaborn Docs")
plt.show()
```

Example 3: Boxplot

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker")
plt.title("Boxplot Example from Seaborn Docs")
plt.show()
```
