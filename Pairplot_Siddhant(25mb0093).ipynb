```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

Showing the dataset

```python
df = pd.read_csv("penguins.csv")
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
```

Removing the Duplicate Dataset

```python
df = df.dropna()
```

Categories for the Plots

```python
features = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
```

The Pair Plot

```python
sns.set(style="ticks")

pairplot = sns.pairplot(
    df[features + ["species"]],
    hue="species",
    kind="scatter",
    diag_kind="kde",
    palette="coolwarm",
    corner=True,
    markers=["o", "s", "D"],
    plot_kws={'alpha':0.6, 'edgecolor':'black'}
)

pairplot.fig.suptitle("Penguins Dataset - Pairplot Analysis", y=1.02, fontsize=16)

plt.show()
```

The Interpretaions

**Main interpretations:**

1.   Adelie penguins:

    *   Smaller body mass (around 30004000 g)
    *   Shorter flippers (around 180 mm)
    *   Distinct cluster at lower values.





2.   Chinstrap penguins:

    *   Similar body mass to Adelie but with longer bills.
    *   Overlaps with Adelie in some features, so harder to separate fully.



3.   Gentoo penguins:

    *   Heaviest (50006000 g).
    *   Longest flippers (~220 mm).
    *   Clearly separated from the other two species.

***Summary:***

This pairplot helps us compare different penguin species based on their body features. We can see that Gentoo penguins are much larger and heavier, Adelie penguins are smaller, and Chinstrap penguins fall in between. The scatterplots also show clear groupings, which means these features can be used to classify the penguin species.
